```python
def rolling_averages(group,cols,new_cols):
  rolling_stats=group[cols].rolling(3,closed='left').mean()
  group[new_cols]=rolling_stats
  group=group.dropna(subset=new_cols)
  return group
cols=['gf', 'ga','xg', 'xga', 'poss', 'attendance','sh_x', 'sot',
       'dist', 'fk', 'pk', 'pkatt', 'sot%', 'g/sh', 'g/sot', 'sota', 'saves',
       'save%', 'cs', 'psxg', 'cmp', 'att_x', 'cmp%', 'totdist', 'prgdist_x',
       'kp', 'ppa', 'prgp', 'tkl','tkl.1','tklw', 'blocks', 'sh_y', 'pass',
       'int', 'err', 'tkl+int', 'touches', 'att pen', 'att_y', 'carries',
       'prgdist_y', 'cpa', 'rec', 'prgr', 'fls', 'fld','crs']
new_cols=[f'{c}_rolling' for c in cols]
rolling_averages(group,cols,new_cols)
```

```python
group=group.dropna(subset=new_cols)
group.columns
```

```python
df_rolling=df.groupby('team').apply(lambda x:rolling_averages(x,cols,new_cols))
df_rolling
```

```python
df_rolling=df_rolling.droplevel('team')
df_rolling.columns
```

```python
df_rolling['goal_conversion_rate'] = df_rolling['gf_rolling'] / df_rolling['sh_x_rolling']
df_rolling['save_percentage'] = df_rolling['saves_rolling'] / (df_rolling['saves_rolling'] + df_rolling['ga_rolling'])
```

```python
custom_weights={2017: 1.0, 2018: 1.1, 2019: 1.2, 2020: 1.5, 2021: 1.8, 2022: 2.0, 2023: 2.5, 2024: 3}
```

```python
import numpy as np
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
train=df_rolling[df_rolling['date']<'2024-01-01']
test=df_rolling[df_rolling['date']>'2024-01-01']
kk=predictors+new_cols
print(len(kk))
train_weights=train['season'].map(custom_weights)
```

```python
rf=RandomForestClassifier(n_estimators=50,min_samples_split=10,random_state=1)
```

```python
rf.fit(train[kk],train['result_code'],sample_weight=train_weights)
preds=rf.predict(test[kk])
```

```python
from sklearn.metrics import f1_score, recall_score, precision_score
```

```python
f1_score(test['result_code'],preds,average='micro')
```

```python
recall_score(test['result_code'],preds,average='micro')
```

```python
precision_score(test['result_code'],preds,average='micro')
```

```python
predictions = pd.DataFrame({
    'predicted_result': preds,
    'actual_result': test['result_code'],
    'date': test['date'],
    'team': test['team'],
    'opponent': test['opponent'],
    'result': test['result'],
    'season': test['season']
})
```
